```python
from pyspark import SparkContext

sc = SparkContext("local", "nothing")
```

```python

list1 = (9,19,29,39,49,59,69,79,89,99)

print(type(list1))

rdd1 = sc.parallelize(list1)

print(type(rdd1))

# create new rdd  as >=5 from rdd1

rdd2 = rdd1.filter( lambda x: x>=3)

rdd2.collect()


# create new rdd by adding 10 to each value in rdd1

rdd3 = rdd1.map( lambda a : a+69)
rdd3.collect()
```
